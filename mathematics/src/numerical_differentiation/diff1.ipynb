# Numerical Differentiation


```python
from math import exp, sin, cos
import matplotlib.pyplot as plt
```

```python
def f(x):
    return exp(sin(2 * x))


def fprime(x):
    return 2 * exp(sin(2 * x)) * cos(2 * x)


def calc_fd(f, x, h):
    fd = (f(x + h) - f(x)) / h
    return fd


def calc_cd(f, x, h):
    cd = (f(x + h / 2) - f(x - h / 2)) / h
    return cd
```

```python
x = 0.5
an = fprime(x)
hs = [10 ** (-i) for i in range(1, 12)]
fds = [abs(calc_fd(f, x, h) - an) for h in hs]
cds = [abs(calc_cd(f, x, h) - an) for h in hs]
rowf = "{0:1.0e} {1:1.16f} {2:1.16f}"
print("h abs. error in fd abs. error in cd")

for h, fd, cd in zip(hs, fds, cds):
    print(rowf.format(h, fd, cd))
```

```python
# print fds and cds in a plot
plt.loglog(hs, fds, "o-", label="fd", marker="o", linestyle="--")
plt.loglog(hs, cds, "o-", label="cds", marker="x", linestyle="--")
```

```python
plt.plot(hs, cds, "o-", label="cds")
plt.ylabel("y")
plt.grid(True, which="both", ls="-")
plt.legend()
```

```python
fig, ax = plt.subplots()
ax.loglog(hs, fds, marker="o", linestyle="--")
ax.legend(fontsize=12)

ax.loglog(hs, cds, marker="x", linestyle="--")
ax.legend(fontsize=12)

ax.set_xlabel("fds")
ax.set_ylabel("cds")
ax.grid(True, which="both", ls="-")
ax.legend(fontsize=12)

plt.show()
```
