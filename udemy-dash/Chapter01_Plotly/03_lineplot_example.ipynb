```python
import pandas as pd
import plotly.offline as pyo
import plotly.graph_objs as go
```

```python
df = pd.read_csv('../data/2010YumaAZ.csv')
df
# print(df.info())
# print(df.columns)
# print(df.index)
```

```python
df.drop('LST_DATE', axis=1, inplace=True)
df.set_index(df["DAY"], inplace=True)
df.drop("DAY", axis=1, inplace=True)
df
```

```python
traces = [go.Scatter(
    x=df["LST_TIME"].loc[day],
    y=df["T_HR_AVG"].loc[day],
    mode='markers+lines',
    name=day
) for day in df.index.unique()]

# data = [{
#     "x": df["LST_TIME"],
#     "y": df[df["DAY"] == day]["T_HR_AVG"],
#     "name": day
# } for day in df.index.unique()]


layout = go.Layout(title="Average Temperature from June 1-7 in Yuma, Arizona")

fig = go.Figure(data=traces, layout=layout)

pyo.plot(fig)
```
